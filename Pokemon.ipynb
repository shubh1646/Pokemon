```python
import os
from pathlib import Path
from  keras.preprocessing import image
import matplotlib.pyplot as plt 
```

```python
p = Path("./Dataset")
dirs = p.glob("*")

image_data = []
labels_dict ={"Diglett":0,"Pikachu":1,"Hitmonchan":2,"Ditto":3,"Bulbasaur":4,"Caterpie":5,"Hitmonlee":6,"Charizard":7}
label2pokemon = {0:"Diglett",1 : "Pikachu",2 : "Hitmonchan",3 : "Ditto", 4 : "Bulbasaur", 5 :"Caterpie", 6 :"Hitmonlee",7 : "Charizard"}
labels = []
for folder_dir in dirs:
    label  = str(folder_dir).split("/")[-1]
    print(folder_dir)
    count = 0;
    
    for img_path in folder_dir.glob("*.jpg"):
        img   = image.load_img(img_path,target_size=(100,100))
        img_array = image.img_to_array(img)
        image_data.append(img_array)
        labels.append(labels_dict[label])
        count+=1
        
    print(count)
    
    
    
    
    
```
